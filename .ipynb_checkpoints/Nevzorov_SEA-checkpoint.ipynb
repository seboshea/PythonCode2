```python
from PICASSO_functions import DataPaths, LoadFlightData2Dict
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
```

Load flight C081 data

```python
PathDict=DataPaths()
FlightDict_C081=LoadFlightData2Dict(1,0,1,0,'C081',PathDict)
```

```python
TWC_021=FlightDict_C081['TWC_021']
TWC_083=FlightDict_C081['TWC_083']
TWC_g_m3=FlightDict_C081['TWC_g_m3']
LWC_021=FlightDict_C081['LWC_021']
LWC_083=FlightDict_C081['LWC_083']
LWC_g_m3=FlightDict_C081['LWC_g_m3']
IWC_021=FlightDict_C081['IWC_021']
IWC_083=FlightDict_C081['IWC_083']
IWC_g_m3=FlightDict_C081['IWC_g_m3']
CDP_gm3=FlightDict_C081['CDP_gm3']
TAT_ND_R_C=FlightDict_C081['TAT_ND_R_C'] 
Time_Core=FlightDict_C081['Time_Core']
```

Scatter plots 

```python
fig=plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.plot(TWC_g_m3,TWC_083, 'o', markersize=1 , label='TWC_083')
plt.plot(TWC_g_m3,TWC_021, 'o', markersize=1 , label='TWC_021')
plt.xlim([-0.05, 1])
plt.ylim([-0.05, 1])
plt.xlabel('TWC Nevzorov, g m$^{-3}$')
plt.ylabel('TWC, g m$^{-3}$')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(LWC_g_m3,LWC_083, 'o', markersize=1, label='LWC_083')
plt.plot(LWC_g_m3,LWC_021, 'o', markersize=1, label='LWC_021' )
plt.plot(LWC_g_m3,CDP_gm3, 'o', markersize=1, label='CDP' )
plt.xlim([-0.05, 0.7])
plt.ylim([-0.05, 0.7])
plt.xlabel('LWC Nevzorov, g m$^{-3}$')
plt.ylabel('LWC, g m$^{-3}$')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(IWC_g_m3,IWC_083, 'o', markersize=1, label='IWC_083')
plt.plot(IWC_g_m3,IWC_021, 'o', markersize=1, label='IWC_021' )

plt.xlim([-0.05, 0.7])
plt.ylim([-0.05, 0.7])
plt.xlabel('IWC Nevzorov, g m$^{-3}$')
plt.ylabel('IWC, g m$^{-3}$')
plt.legend()
```

Time Series

```python
    fig=plt.figure(figsize=(10,10)) 
    plt.rcParams.update({'font.size': 10})
    formatter = DateFormatter('%H:%M')
    
    ax1=fig.add_subplot(3, 1, 1)  
    plt.ylabel('T, Celsius')
    plt.plot(Time_Core, TAT_ND_R_C, '-')
    plt.ylim([-40, 5])
    
    
    plt.subplot(3, 1, 2, sharex=ax1)
    plt.plot(Time_Core, TWC_g_m3,linewidth=0.5, label='Nevzorov')
    plt.plot(Time_Core, TWC_021,linewidth=0.5, label='TWC_021')
    plt.plot(Time_Core, TWC_083,linewidth=0.5, label='TWC_083')
    plt.legend()
    plt.ylabel('TWC, g m$^{-3}$')
    plt.ylim([-0.05, 1])
    
    plt.subplot(3, 1, 3, sharex=ax1)
    plt.plot(Time_Core, LWC_g_m3,linewidth=0.5, label='Nevzorov')
    plt.plot(Time_Core, LWC_021,linewidth=0.5, label='LWC_021')
    plt.plot(Time_Core, LWC_083,linewidth=0.5, label='LWC_083')
    plt.plot(Time_Core, CDP_gm3,linewidth=0.5, label='CDP')
    plt.legend()
    plt.ylabel('LWC, g m$^{-3}$')
    plt.ylim([-0.05, 0.7])
    
    plt.gca().xaxis.set_major_locator(mdates.HourLocator() )
    plt.gca().xaxis.set_minor_locator(mdates.MinuteLocator(30) )

    plt.gcf().axes[0].xaxis.set_major_formatter(formatter)
    plt.xlabel('time') 
   
    plt.xlim(['2018-02-13 05:00:00','2018-02-13 10:00:00'])
    plt.tight_layout()   
    
    
```
